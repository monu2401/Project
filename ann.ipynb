```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
```

```python
data=pd.read_csv("IMDb.csv")
```

```python
my_cols=['castTotalLikes','directorLikes','actor1likes','movieLikes','fbPosters','year','duration','genre','contentRating','criticReviews','userReviews','userVotes','rating']
data=data[my_cols]
# to display top five rows from table.
data.head()
```

```python
from sklearn import preprocessing
Label = preprocessing.LabelEncoder()

data['genre']=Label.fit_transform(data['genre'])
data.head()
```

```python
data['contentRating']=Label.fit_transform(data['contentRating'])
data.head()
```

```python
data=data.dropna()
data.describe().round(1)
```

```python
data.corr().round(3)
```

```python
train,test = train_test_split(data,random_state=42)
X_train = train[train.columns[2:30]]
y_train = train['genre']
X_test = test[test.columns[2:30]]
y_test = test['genre']
```

```python
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neural_network import MLPClassifier
MLP = MLPClassifier(hidden_layer_sizes=(10,10,10),max_iter=1000)
MLP.fit(X_train,y_train.values.ravel())
```

```python
predictions = MLP.predict(X_test)
predictions
```

```python
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
```
